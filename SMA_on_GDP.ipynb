```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas_datareader import data
import datetime
```

```python
start = datetime.datetime(2010, 1, 1)

end = datetime.datetime(2017, 1, 1)

gdp = data.DataReader('GDP', 'fred', start, end)
```

```python
#gdp.ix['2017-01-01']
print(gdp)
```

```python
set=gdp['GDP']
```

```python
def movingaverage(values,window):
    weights=np.repeat(1.0,window)/window
    smas=np.convolve(values,weights,'valid')
    return smas
```

```python
fivestep=movingaverage(set,5)
tenstep=movingaverage(set,10)
```

```python
plt.plot(gdp)
plt.show()
```

```python
plt.plot(fivestep) #SMA using a window of five on gdp
plt.show()
```

```python
plt.plot(tenstep)
plt.show()
```
